```python
 # zoom and then rotate with another way for test 2
import os
import random
from PIL import Image

def generate_random_floats(num_points, min_value, max_value):
    return [round(random.uniform(min_value, max_value), 2) for _ in range(num_points)]

def zoom_in(image, zoom_factor, center_x, center_y, output_size):
    original_width, original_height = image.size
    new_width = int(original_width / zoom_factor)
    new_height = int(original_height / zoom_factor)
    
    left = int(center_x * original_width - new_width / 2)
    top = int(center_y * original_height - new_height / 2)
    right = left + new_width
    bottom = top + new_height
    
    cropped_image = image.crop((left, top, right, bottom))
    resized_image = cropped_image.resize(output_size, Image.ANTIALIAS)
    
    return resized_image

def rotate_image(image, angle):
    return image.rotate(angle)

def save_image(image, directory, filename):
    if not os.path.exists(directory):
        os.makedirs(directory)
    image.save(os.path.join(directory, filename))

def main():
    input_image_path = "test.jpg"  
    num_images = 200  # Adjust the number of output images
    output_directory = "zoomed_rotated_images" 
    output_size = (500, 500)  # Adjust the size of the output images
    
    image = Image.open(input_image_path)


    generated_rotations = set()

    for i in range(num_images):
        zoom_factor = random.uniform(1.1, 2.0)  
        center_x = random.uniform(0.2, 0.8)  
        center_y = random.uniform(0.2, 0.8)  

      
        rotation_angle = None
        while rotation_angle is None or rotation_angle in generated_rotations:
            rotation_angle = round(random.uniform(-40, 40), 2)

       
        generated_rotations.add(rotation_angle)

        zoomed_image = zoom_in(image, zoom_factor, center_x, center_y, output_size)
        rotated_image = rotate_image(zoomed_image, rotation_angle)
        
        output_filename = f"zoomed_rotated_image_{i}.png"
        save_image(rotated_image, output_directory, output_filename)

if __name__ == "__main__":
    main()
```
